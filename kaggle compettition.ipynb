```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
import statistics
import warnings
warnings.filterwarnings('ignore')
```

```python
train=pd.read_csv("E:\\Practice datasets\\house-prices-advanced-regression-techniques\\train.csv")
test=pd.read_csv("E:\\Practice datasets\\house-prices-advanced-regression-techniques\\test.csv")
```

```python
train[train.select_dtypes(['object']).columns]=train.select_dtypes(['object']).apply(lambda x: x.astype('category'))
train
```

```python
train['LotFrontage']=train['LotFrontage'].fillna(train['LotFrontage'].mean())
train['MasVnrType']=train['MasVnrType'].fillna(statistics.mode(train['MasVnrType']))
train['MasVnrArea']=train['MasVnrArea'].fillna(statistics.median(train['MasVnrArea']))
```

```python
c = train.select_dtypes(np.number).columns
train[c] = train[c].fillna(train[c].mean())
d = test.select_dtypes(np.number).columns
test[d] = test[d].fillna(test[d].mean())
```

```python
cols =train.select_dtypes('category').columns
train[cols]=train[cols].fillna(train.mode().iloc[0])
cols1 =test.select_dtypes('category').columns
test[cols1]=test[cols1].fillna(test.mode().iloc[0])
```

```python
cat_col=train.select_dtypes('category').columns
num_col=train.select_dtypes(exclude=['category'])
```

```python
colums=train.columns
colums
```

```python
test=pd.DataFrame(test,columns=colums)
```

```python
from sklearn.preprocessing import LabelEncoder
lb=LabelEncoder()
for i in cat_col:
    train[i]=lb.fit_transform(train[i])
    test[i]=lb.fit_transform(test[i])
```

```python
train1=train.drop(columns=['Id','SalePrice'])
test=test.drop(columns=['Id','SalePrice'])
```

```python
import xgboost as xg
```

```python
xgb_r = xg.XGBRegressor(objective ='reg:linear',
                  n_estimators = 10, seed = 123)
 
# Fitting the model
xgb_r.fit(train1,y_train)
 
# Predict the model
pred = xgb_r.predict(test)
 
```
